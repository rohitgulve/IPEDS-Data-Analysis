```python
# Step 1: Calculate the percentage of missing values in each column
percent_missing = IPED_data.isna().sum() / len(IPED_data) * 100

# Step 2: Filter columns with more than 20% missing values
high_missing_columns = percent_missing[percent_missing > 20]

# Step 3: Sort the columns by percentage of missing values (highest to lowest)
high_missing_columns = high_missing_columns.sort_values(ascending=False)

# Step 4: Plot the results as a bar chart
ax = high_missing_columns.plot.bar(
    title='Percentage of Missing Values in Columns',  # Title of the chart
    figsize=(12, 5)  # Size of the chart
)

# Step 5: Add a label for the y-axis
ax.set_ylabel('% of Missing Values')
```

```python
missing_percentage = (IPED_data.isnull().sum() / len(IPED_data)) * 100
print(missing_percentage)
```

```python
missing_percentage = missing_percentage.sort_values(ascending=False)
print(missing_percentage)
```

```python
threshold = 0.2  # 20% missing values

# Calculate the percentage of missing values for each column
missing_percentage = IPED_data.isnull().mean()
missing_percentage
```

```python
# Identify columns with more than the threshold percentage of missing values
columns_to_drop = missing_percentage[missing_percentage > threshold].index
```

```python
columns_to_drop
```

```python
# Drop the identified columns
df_cleaned = IPED_data.drop(columns=columns_to_drop)
```

```python
# Replacing extra spaces and special characters, ensuring only one underscore
df_IPED_cleaned.columns = (
    df_IPED_cleaned.columns
    .str.replace(r'\s+|\/|\s-\s|-', '_', regex=True)  # Replace spaces, '/', ' - ', and '-' with '_'
    .str.replace(r'_+', '_', regex=True)  # Replace multiple underscores with a single underscore
    .str.lower()  # Convert to lowercase
    .str.replace(r"[\',-]", "", regex=True)  # Remove special characters like commas, apostrophes, and hyphens
)
```

```python
df_IPED_cleaned.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Define the columns and their corresponding titles
columns = ['applicants_total', 'admissions_total', 'enrolled_total']
titles = [
    "Histogram of Number of Applications",
    "Histogram of Number of Admissions",
    "Histogram of Number of Enrollments"
]

# Create a figure with a specific size
plt.figure(figsize=(16, 6))

# Loop through the columns and titles
for i, (col, title) in enumerate(zip(columns, titles), 1):
    plt.subplot(1, 3, i)
    sns.histplot(df_IPED_cleaned[col], bins=50, kde=True, color='skyblue')
    plt.title(f"{title}\nMean: {df_IPED_cleaned[col].mean():.1f}, Median: {df_IPED_cleaned[col].median():.1f}")
    plt.xlabel(f'Number of {col.replace("_total", "").replace("_", " ").capitalize()}')
    plt.ylabel('Frequency')
    plt.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout(pad=2)
plt.show()
```

```python
# Create a figure with a specific size
plt.figure(figsize=(16, 6))

# First subplot: Applications vs Admissions
plt.subplot(1, 2, 1)
plt.title('APPLICATIONS VS ADMISSIONS')
sns.scatterplot(
    y=df_IPED_cleaned['admissions_total'], 
    x=df_IPED_cleaned['applicants_total'], 
    hue=df_IPED_cleaned['control_of_institution']
)
plt.ylabel('Number of Admissions')
plt.xlabel('Number of Applications')
plt.grid()

# Second subplot: Admissions vs Enrollments
plt.subplot(1, 2, 2)
plt.title('ADMISSIONS VS ENROLLMENTS')
sns.scatterplot(
    x='admissions_total', 
    y='enrolled_total', 
    data=df_IPED_cleaned, 
    hue='control_of_institution'
)
plt.ylabel('Number of Enrollments')
plt.xlabel('Number of Admissions')
plt.grid()

# Adjust layout to prevent overlap
plt.tight_layout(pad=2)
plt.show()
```

```python
# Calculate enrollment rate as a percentage and round to 2 decimal places
df_IPED_cleaned['enrollment_rate'] = (
    (df_IPED_cleaned['enrolled_total'] / df_IPED_cleaned['admissions_total']) * 100
).round(2)

df_IPED_cleaned.head(2)
```

```python
# Calculate acceptance rate as a percentage and round to 2 decimal places
df_IPED_cleaned['acceptance_rate'] = (
    (df_IPED_cleaned['admissions_total'] / df_IPED_cleaned['applicants_total']) * 100
).round(2)
```
